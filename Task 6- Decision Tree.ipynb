In this Notebook we will apply Dicision Tree Algorithm on Iris Data to make a model that would be able to predict the Species if we feed new data to the model.

```python
# Importing required libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Importing the dataset as pandas Dataframe
file_loc = 'C:/Users/Prometheus/Untitled Folder/datasets/Iris.csv'
iris = pd.read_csv(file_loc)
iris.head(10)
```

We will separate Species and features columns for the purpose for training the model

```python
species = iris['Species']
df = iris.drop(columns=['Id', 'Species'])
df.head()
```

```python
dt = DecisionTreeClassifier()
dt.fit(df,species) # Fit Decision Tree Classifier model to data with species as target variable.
```

```python
import graphviz 
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dt, out_file=dot_data,feature_names=df.columns,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graphviz.Source(graph)  
Image(graph.create_png())
```
